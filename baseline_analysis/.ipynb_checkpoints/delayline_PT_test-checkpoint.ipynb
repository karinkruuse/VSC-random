```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch
```

```python
data = np.load("measurement_data/DelayLineTest_20260127_134423.npy", mmap_mode="r")
print(data.shape, data.dtype)
```

```python
t = data['Time (s)']

#### Delayed signal
f1     = data['Input 1 Frequency (Hz)'] # Hz
phi1   = data['Input 1 Phase (cyc)'] # cyc

#### DDS signal
f2     = data['Input 2 Frequency (Hz)'] # Hz
phi2   = data['Input 2 Phase (cyc)'] # cyc
f2_mean = np.mean(f2)

#### No dealy signal
f3     = data['Input 3 Frequency (Hz)'] # Hz
phi3   = data['Input 3 Phase (cyc)'] # cyc
```

```python
# Sanity check, from conf 150 Hz -ish
fs = 1 / np.mean(np.diff(t))
print("Sampling frequency (Hz):", fs)
```

```python
set_delay = 8.1 # s
```

```python
def asd_welch(x, fs, nperseg=2**14, noverlap=None, window="hann",
              detrend="constant", remove_mean=True, return_psd=False):
    x = np.asarray(x); 
    if remove_mean: x = x - np.mean(x)
    if noverlap is None: noverlap = nperseg//2
    f, P = welch(x, fs, window=window, nperseg=nperseg,
                 noverlap=noverlap, detrend=detrend, scaling="density")
    A = np.sqrt(P)
    return (f, A, P) if return_psd else (f, A)
```

```python
def plot_asd(x, fs, nperseg=2**14, fmin=None, fmax=None, label=None):
    f, P = welch(x - np.mean(x), fs, nperseg=nperseg, noverlap=nperseg//2)
    A = np.sqrt(P)
    m = (A > 0)
    if fmin is not None: m &= f >= fmin
    if fmax is not None: m &= f <= fmax
    plt.grid()
    plt.loglog(f[m], A[m], label=label)
    plt.xlabel("Hz"); plt.ylabel("ASD [units/√Hz]")
    if label: plt.legend()
```

```python
plot_asd(phi1, fs, fmin=1e-3, label="φ1")
plot_asd(phi2, fs, fmin=1e-3, label="φ₂")
plot_asd(phi3, fs, fmin=1e-3, label="φ3")
plt.ylabel("ASD [cycles/√Hz]")
plt.show()
```

```python
def delay(x, t, tau):
    y = np.interp(t - tau, t, x, left=np.nan, right=np.nan)
    m = np.isfinite(y)
    return t[m], y[m]
```
